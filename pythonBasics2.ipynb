```python

def word_count(string):
    dicto={}
    for i in range (len(string)):
        if string[i] in ".,?!:;-()/'":
            string = string.replace(string[i], " ")
    
        
    words=string.split()
    for word in words:
        if word in dicto.keys():
            dicto[word]+=1
        else:
            dicto[word]=1
    return dicto


word_count("ahmed sayed zizo !!!!")
```
