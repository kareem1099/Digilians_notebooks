```python
"""
- Create `bonus` column = 10% of salary_filled
- Create `total_compensation` = salary_filled + bonus
- Create `experience_level`:
  - Junior: < 4 years
  - Mid: 4–6 years
  - Senior: > 6 years
- Use `apply()` to categorize `salary`:
  - Low: < 65,000
  - Medium: 65,000–75,000
  - High: > 75,000
"""
emp_df["bonus"]=emp_df["salary"]*10/100
emp_df["total_compensation"]=emp_df["salary"]+emp_df["bonus"]


conditions=[
    (emp_df["years_experience"]<4),
    (emp_df["years_experience"]>4)&(emp_df["years_experience"]<6),
    (emp_df["years_experience"]>6)

]

y=["Junior","Mid","Senior"]

import numpy as np

emp_df["experience_level"]=np.select(conditions,y,default="unknown")
emp_df

def categorize(salary):
    if salary<=65000:
        return"low"
    if salary>=65000 and salary<=75000 :
        return"Medium"
    if salary<=75000:
        return"High"
    
emp_df["cat_salary"]=emp_df["salary"].apply(categorize)
emp_df
```
