8. Mini Program — Shopping Cart
Write a program:
 List of prices
 Calculate total
 If total > 100 → print “Discount Applied”

```python
Cart=[]
x= int(input("How many items did you buy"))
for i in range(x):
    y=int(input(f"enter the price of item number {i+1}"))
    Cart.append(y)



if sum(Cart)>100:
    print("Discount Applied")
```
