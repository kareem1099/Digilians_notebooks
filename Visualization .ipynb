###  Exercise 6 — Box Plot with Seaborn

Create a **box plot** using Seaborn that compares the **math scores** of students grouped by their **grade** (A, B, C).

Requirements:
- X-axis → grade
- Y-axis → math
- Use `hue='grade'` to color by grade

###  Exercise 7 — Heatmap with Seaborn

Create a **correlation heatmap** using Seaborn for the columns: `math`, `english`, `science`, and `study_hours`.

Requirements:
- Calculate the correlation matrix using `.corr()`
- Use `annot=True` to show values on the heatmap
- Use `cmap='coolwarm'`
- Title: `'Correlation Between Subjects and Study Hours'`

**After the chart:** Answer — which two variables have the **strongest** correlation?

````python
#  Write your code here

corr_matrix=df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Between Subjects and Study Hours")
plt.show()
````

>  **Strongest correlation is between:** Maths and study hours

###  Exercise 9 — Interactive Scatter Plot with Plotly

Create an **interactive scatter plot** using Plotly Express that shows **study_hours vs math scores**.

Requirements:
- X-axis → study_hours
- Y-axis → math
- Color each point by `grade`
- Show student `name` on hover using `hover_name='name'`
- Title: `'Study Hours vs Math Score by Grade'`

````python

fig = px.scatter(
    df,
    x="study_hours",
    y="math",
    color="grade",
    hover_name="name",
    title="Study Hours vs Math Score by Grade"
)

fig.show()
````

###  Exercise 10 — Grouped Bar Chart with Plotly

Create an **interactive grouped bar chart** using Plotly Express that shows **math, english, and science scores** for each student.

**Hint:** You need to reshape the data using `pd.melt()` first:
```python
df_melted = pd.melt(df, id_vars='name', value_vars=['math', 'english', 'science'],
var_name='subject', value_name='score')
```

Requirements:
- X-axis → name
- Y-axis → score
- Color → subject
- Use `barmode='group'`
- Title: `'Scores by Subject per Student'`

````python
df_melted = pd.melt(df, id_vars='name', value_vars=['math', 'english', 'science'],
var_name='subject', value_name='score')
fig = px.bar(
    df_melted,          
    x="name",
    y="score",
    color="subject",    
    barmode="group",
    title="Scores by Subject per Student"
)

fig.show()
````

##  Bonus Challenge (Optional)

Create **any chart of your choice** using the student dataset that tells an interesting story.

Requirements:
