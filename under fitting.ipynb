```python
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python

data=pd.read_csv("C:/Users/Admin/Downloads/possum.csv")
```

```python
data
```

```python
data.info()
```

```python
for col in list(data.columns):
    print(data[col].unique())
```

```python
data["age"].value_counts()

data.drop(["case","site","Pop","earconch","eye"],axis=1,inplace=True)
```

```python
mean_val=data.age.mean()
data.age.fillna(mean_val,inplace=True)
```

```python
mean_val=data.footlgth.mean()
data.footlgth.fillna(mean_val,inplace=True)
```

```python
data.sex= data.sex.map({"m": 1, "f": 0})
data["Pop"]=data["Pop"].map({"Vic":1,"other":0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=LinearRegression()

model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
trained=model.predict(x_train)
predicted=model.predict(x_test)
r2_sc_train=r2_score(y_train,trained)
r2_sc_test=r2_score(y_test,predicted)
print(r2_sc_train)
print(r2_sc_test)
```
